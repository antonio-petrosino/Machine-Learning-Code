Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values #location of column by indices, saltiamo colonna labellata
# la parentesi [] include il primo indice, ma esclude il secondo indice
y = dataset.iloc[:, -1].values
```

training the linear regr model whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

training the polynomial regression whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X) #questo trasforma solo la matrice delle features

lin_reg_2 = LinearRegression() # in ogni caso serve questa regres
lin_reg_2.fit(X_poly, y)
```

 visual linear reg. result

```python
plt.scatter(X, y, color = 'red') # prediction
plt.plot(X, lin_reg.predict(X), color = 'blue') # per la normale linare
plt.title('Truth of Bluff (Linear Regression)')
plt.xlabel('Position Level [#]')
plt.ylabel('Salary [$]')
plt.show()
```

visual poly reg. result

```python
plt.scatter(X, y, color = 'red') # prediction
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue') # per la normale linare
plt.title('Truth of Bluff (Polynomial Regression)')
plt.xlabel('Position Level [#]')
plt.ylabel('Salary [$]')
plt.show()
```

visual poly regr (for higher resolution and smoother curve)

```python
#poly_reg = PolynomialFeatures(degree = 4) # change this degree
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
# non interpoliamo solo ad X interi ma mettiamo degli X intermedi più piccoli (uno ogni 0.1)

plt.scatter(X, y, color = 'red') # prediction
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue') # per la normale linare
plt.title('Truth of Bluff (Polynomial Regression)')
plt.xlabel('Position Level [#]')
plt.ylabel('Salary [$]')
plt.show()
```

predicting a new result with linear regression

```python
lin_reg.predict([[6.5]]) # [[ ]] solo una parentesi crea un vettore, la doppia crea un array
```

predicting a new result with poly regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]])) # [[ ]] solo una parentesi crea un vettore, la doppia crea un array
# che equivale a [[6.5**0, 6.5**1, 6.5**2, 6.5**3, 6.5**4]]
```
