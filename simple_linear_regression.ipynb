# Simple Linea Regression

Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Import dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Split train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

Training phase

```python
from sklearn.linear_model import  LinearRegression
custom_regressor = LinearRegression()
custom_regressor.fit(X_train, y_train)
```

Predicting the test results

```python
y_pred = custom_regressor.predict(X_test) # mi restituisce y previsto: y_test è quello reale, y_pred è quello previsto
```

Visualizing training result

```python
plt.scatter(X_train, y_train, color='red') # plotto i punti rossi reali del training set
plt.plot(X_train, custom_regressor.predict(X_train), color = 'blue') # plotto la linea di regressione lineare
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of experience [#]')
plt.ylabel('Salary [$]')
plt.show()
```

Visualizing test result

```python
plt.scatter(X_test, y_test, color='red') # plotto i punti rossi reali del training set
#plt.plot(X_train, custom_regressor.predict(X_train), color = 'blue') # plotto la linea di regressione lineare
plt.plot(X_test, y_pred, ':',color = 'green') # plotto la linea di regressione lineare
# non c'è bisogno di riportare y_red perchè è la stessa identica retta
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of experience [#]')
plt.ylabel('Salary [$]')
plt.show()
```
